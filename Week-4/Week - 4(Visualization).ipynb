Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Loading Datasets

```python
metadata_country = pd.read_csv("/content/Metadata_Country.csv")
country_population = pd.read_csv("/content/country_population.csv")
fertility_rate = pd.read_csv("/content/fertility_rate.csv")
life_expectancy = pd.read_csv("/content/life_expectancy.csv")
```

To prevent data loss, create copy of dataset.

```python
country = metadata_country.copy()
population = country_population.copy()
fertilityRate = fertility_rate.copy()
lifeExpectancy = life_expectancy.copy()
```

# **Data Cleaning**

---



# **Working with population dataset**




```python
#Dropping columns Indicator Name and Indicator Code

population.drop(columns=['Indicator Name','Indicator Code'], inplace=True)
```

```python
# Checking for Null values in population dataset

population.isna().sum()
```

```python
population.info()
```

```python
# Replacing NAN values with mean value

years = [str(year) for year in range(1960, 2017)]
population[years] = population[years].round(decimals = 1)
population[years] = population[years].astype('float64')

for year in years:
    population[year].fillna(population[year].mean(), inplace=True)
population.head()
```

```python
# Again checking for Null values in population dataset

population.isna().sum()
```

```python
population.shape
```

```python
# Using melt() function which is used to change the DataFrame format from wide to long.

m_population = pd.melt(population,
        id_vars='Country Code',
        value_vars=years,
        var_name='Year',
        value_name='Population')
m_population.shape  # Getting number of rows and columns
```

```python
m_population
```

# **Working with country dataset**

```python
# Dropping columns IncomeGroup SpecialNotes and TableName

country.drop(columns=['IncomeGroup','SpecialNotes','TableName'], inplace=True)
```

```python
# Checking for Null values in country dataset

country.isna().sum()
```

```python
country.drop(columns=['Unnamed: 5'], inplace=True)
```

```python
country.info()
```

```python
country.shape # Getting number of rows and columns
```

```python
country
```

```python
# Merging m_population and country dataset

country_population = pd.merge(country, m_population,how='left',on='Country Code')
country_population
```

# **Data visualization**

---



# Visualizing Fertility vs Life Expectancy across different regions




```python
fig = px.scatter(merge_dataset,
                 x='fertilityRate',
                 y='lifeExpectancy',
                 size='Population',
                 color='Region',
                 hover_name='Country Code',
                 animation_frame='Year',
                 animation_group='Country Code',
                 template='plotly_dark',
                 size_max=100,
                 range_x=[0, 10],
                 range_y=[10, 90],
                 title='Animated Bubble Chart: Fertility vs Life Expectancy By Region',
                 labels={'fertilityRate': 'fertilityRate', 'lifeExpectancy': 'lifeExpectancy', 'Population': 'Population', 'Region': 'Region', 'Country Code': 'Country Code'})

fig.update_layout(
    xaxis_title='Fertility',
    yaxis_title='Life Expectancy',
    legend_title='Region',
    hoverlabel=dict(
        font_size=15,
        font_family='Avenir Next LT Pro, Times New Roman'))

fig.show()
```

# Visualizing average Fertility rate trends across different regions

```python
avg_fertilityRate = merge_dataset.groupby('Region')['fertilityRate'].mean().reset_index()

plt.style.use('dark_background')
plt.figure(figsize=(15, 5))
plt.bar(avg_fertilityRate ['Region'], avg_fertilityRate ['fertilityRate'])
plt.title('Average Fertility Rate By Region')
plt.xlabel('Region')
plt.ylabel('Average Fertility Rate')
plt.xticks(rotation=45)
plt.grid(True, color='grey', linestyle='--', linewidth=0.4)
plt.show()
```

# Correlation Analysis

```python
numerical_columns = merge_dataset.select_dtypes(include=['float64', 'int64']).columns

correlation_matrix = merge_dataset[numerical_columns].corr()

sns.set(style="white")

plt.figure(figsize=(12, 5))

cmap = sns.color_palette("YlGnBu", as_cmap=True)
sns.heatmap(correlation_matrix, cmap=cmap, vmax=.3, center=0,
            linewidths=.5, cbar_kws={"shrink": .5},
            annot=True, fmt=".2f", annot_kws={"size": 10, "color": 'black'})

plt.title('Correlation Heatmap of Numerical Columns')

plt.show()
```
