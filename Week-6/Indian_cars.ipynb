# **Data Cleaning**

---



```python
# Drop column

car_df = car_df.drop('Unnamed: 0', axis = 1)
```

```python
# Rename Make column with Company column

car_df = car_df.rename(columns = {'Make':'Company'})
```

```python
# Check null values in Company column

car_df['Company'].isnull().value_counts()
```

```python
# Drop null values in Company column

car_df = car_df.dropna(subset = ['Company'])
car_df['Company'].isnull().value_counts()
```

```python
# Rename Ex-Showroom_Price with Price

car_df = car_df.rename(columns = {'Ex-Showroom_Price' : 'Price'})
```

```python
# Working with Price column

car_df['Price'] = car_df['Price'].str.replace('Rs.','')
car_df['Price'] = car_df['Price'].str.replace(',','')
car_df['Price']
```

```python
car_df['Price'] = car_df['Price'].astype(int)
```

```python
# Working with Displacement column

car_df['Displacement'].isnull().value_counts()
```

```python
car_df = car_df.dropna(subset=['Displacement'])
car_df['Displacement'].isnull().value_counts()
```

```python
# Working with Cylinders column

car_df['Cylinders'].isna().sum()
```

```python
car_df['Cylinders'] = car_df['Cylinders'].replace(np.nan, car_df.Cylinders.median())
```

```python
car_df['Cylinders'] = car_df['Cylinders'].astype(int)
```

```python
car_df['Cylinders'].isna().sum()
```

# **Data Visualization**

## Univariate analysis

---



# Histogram

```python
plt.ticklabel_format(style = 'plain')
sns.histplot(car_df['Price'], kde = True, color = 'red')
# Adding labels and title
plt.xlabel('Price')
plt.title('Histogram')

# Display the plot
plt.show()
```

```python
count_cars = car_df.Company.value_counts()
count_cars
```

# Countplot

```python
plt.figure(figsize=(20,5))
sns.countplot(data= car_df, x='Company', order = car_df['Company'].value_counts().index, palette ='crest')
plt.xlabel('Car Companies')
plt.ylabel('Count')
plt.title('Number of cars by company')
plt.xticks(rotation = 45)
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10, 4))
sns.histplot(car_df['Fuel_Tank_Capacity'], kde=True, color = 'mediumslateblue')

# Adding labels and title
plt.xlabel('Fuel Tank Capacity')
plt.ylabel('Frequency')
plt.title('Distribution of Fuel Tank Capacity')

# Display the plot
plt.show()
```

# Pie Charts

```python
count = car_df['Company'].value_counts()
top_company = count.head(5)

model_count = car_df['Model'].value_counts()
top_model = model_count.head(5)

fig, axes = plt.subplots(1,2, figsize=(10,5))

axes[0].pie(top_company , labels = top_company.index, autopct = '%1.3f%%', colors= sns.color_palette('Set2'), startangle=90)
axes[0].set_title('Top 5 Car Company')

axes[1].pie(top_model , labels = top_model.index, autopct = '%1.3f%%',colors= sns.color_palette('Set2'), startangle=90)
axes[1].set_title('Top 5 Car Models')
plt.show()
```

Maruti Suzuki dominates the Indian car market with a market share of over 26%.
Hyundai and Mahindra are close competitors, each holding around 20% of the market share.
Tata Motors has a significant presence in the market with a market share of over 16%.
Toyota, a global leader, has a smaller market share in India compared to other Asian carmakers. <br>
The top 5 car companies together control over 80% of the Indian car market.

The top car model is Nexon with a market share of 22.222%. This is significantly higher than the market share of the other four models.
The market share is relatively evenly distributed among the other four models. The Compass, Xuv500, and Kuv100 Nxt all have a market share of around 19.4%, while the Amaze has a market share of 18.5%. <br>
The market is competitive. With five car models having a market share of over 18.5%, it is clear that there is no single dominant player in the market.

# Bivariate analysis

# Scatter plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Price', y='Cylinders', data = car_df, hue='Body_Type', palette='plasma', s = 120, alpha = 0.89)
plt.title('Relationship between Price and Cylinders')
plt.xlabel('Price')
plt.ylabel('Cylinders')
plt.legend(title='Body Type')
plt.show()
```


The predominant body types of various cars typically feature engines with cylinder counts ranging from 4 to 8, associated with lower pricing. Hatchbacks, for instance, are equipped with only 2 cylinders, while sports cars can boast a maximum of 16 cylinders.

# Correlation Matrix

```python
num_column = car_df.select_dtypes(include='number')
matrix = num_column.corr()

plt.figure(figsize =(8,6))
sns.heatmap(matrix, annot = True, cmap = 'Greens', fmt = '.2f', linewidths = 0.5)
plt.title('Correlation Matrix')
plt.show()
```

# Box plot with hue

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.boxplot(x = 'Drivetrain', y = 'Price', data = car_df, hue = 'Drivetrain', ax = ax, palette = 'BuPu')
ax.set_title('Box plot')
plt.show()
```

# Violin Plot

Violin Plot is the combination of box plot and distribution plot

```python
plt.figure(figsize=(10,4))
sns.violinplot(x = 'Company', y = 'Price', data = car_df.head(150), palette = 'plasma')
plt.show()
```

# Pair Plot

```python
sns.pairplot(car_df, hue = 'Body_Type', vars = ['Number_of_Airbags','Seating_Capacity','Fuel_Tank_Capacity','City_Mileage'], height = 2)
plt.suptitle('Pair Plot', y = 1.5)
plt.show()
```
