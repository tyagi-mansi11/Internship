# **Data Cleaning**

---



Data Cleaning is performed on Postal Code column as it contains NULL values.

```python
state_with_empty_postal = store_df.loc[store_df['Postal Code'].isnull(), 'State'].unique()

print("States with empty postal codes:", state_with_empty_postal)
```

```python
city_empty_postal = store_df.loc[store_df['Postal Code'].isnull(), 'City'].unique()

print("Cities with empty postal codes:", city_empty_postal)
```

```python
country_empty_postal = store_df.loc[store_df['Postal Code'].isnull(), 'Country'].unique()

print("Countries with empty postal codes:", country_empty_postal)
```

*Only Burlington city of Vermont in United States contains NULL values in Postal Code Column, fill empty values with searched postal code i.e 05401 and change data type from float to int. <br>
(Burlington, Vermont, has several postal codes and the main postal code for Burlington is 05401.)*

```python
store_df['Postal Code'].fillna(05401.0, inplace=True)
store_df['Postal Code'] = store_df['Postal Code'].astype(int)
```

```python
# Again check for null values in all columns
store_df.isna().sum()
```

```python
date_col = ['Order Date', 'Ship Date']

store_df[date_col] = store_df[date_col].apply(lambda col: pd.to_datetime(col, format='%d/%m/%Y', errors='coerce'))

store_df.dtypes
```

# **Data Visualization**




# Univariate analysis

---



```python
# Histogram

plt.figure(figsize=(20,8))
plt.subplot(2, 1, 2)
plt.hist(store_df['Sub-Category'], bins=20, edgecolor='black', alpha=0.7, color='yellowgreen')
plt.title('Histogram for Sub-category')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Pie chart
sales_category = store_df.groupby('Category')['Sales'].sum().reset_index()
plt.pie(sales_category['Sales'], labels = sales_category['Category'], autopct = '%1.3f%%', colors= sns.color_palette('bright'), startangle=90)
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# Time Line Plot

store_df['Order Year'] = store_df['Order Date'].dt.year
yearly_sales = store_df.groupby('Order Year')['Sales'].sum().reset_index()

plt.figure(figsize=(8, 6))
plt.plot(yearly_sales['Order Year'], yearly_sales['Sales'], marker='o', linestyle='-', color='royalblue')
plt.title('Overall Sales Over the Time')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(yearly_sales['Order Year'].astype(int))
plt.show()
```

# Bivariate analysis


---




```python
# Bar Plot

plt.figure(figsize=(20,5))
sns.barplot(data = store_df, x='Sub-Category', y = 'Sales', order = store_df['Sub-Category'].value_counts().index, palette = 'plasma')
plt.title('Sales By Sub-Category')
plt.show()
```

```python
# Line chart

plt.figure(figsize=(8,6))
sns.lineplot(x ='Region', y = 'Sales', hue = 'Category', data = store_df, palette = 'bright')
plt.title('Sales Trends across Region')
plt.legend(title = 'Category')
plt.show()
```

```python
# Pie chart

sales_category = store_df.groupby('Ship Mode')['Sales'].sum().reset_index()
plt.pie(sales_category['Sales'], labels = sales_category['Ship Mode'], autopct = '%1.3f%%', colors = sns.color_palette("colorblind"), startangle=90)
plt.title('Sales Distribution by Shipping Mode')
plt.show()
```

```python
store_df['Month'] = store_df['Order Date'].dt.month
store_df['Year'] = store_df['Order Date'].dt.year
store_df['Quarter'] = store_df['Order Date'].dt.quarter

sales_monthly = store_df.groupby('Month')['Sales'].sum().reset_index()
sales_yearly = store_df.groupby('Year')['Sales'].sum().reset_index()
sales_quarterly = store_df.groupby('Quarter')['Sales'].sum().reset_index()
```

```python
# Line chart

sns.lineplot(data = sales_monthly, x = 'Month', y = 'Sales', marker = 'o', color = 'indigo')
plt.title('Monthly Sales')
plt.grid()
plt.show()
```

```python
sns.lineplot(data = sales_yearly, x = 'Year', y = 'Sales', marker = 'o', color = 'orangered')
plt.title('Yearly Sales')
plt.grid()
plt.show()
```

```python
sns.lineplot(data = sales_quarterly, x = 'Quarter', y = 'Sales', marker = 'o', color = 'forestgreen')
plt.title('Quarterly Sales')
plt.grid()
plt.show()
```

```python
# Customer Segment Analysis

col = ['Segment', 'Sales', 'Year']
col = store_df[col]

plt.figure(figsize=(8,6))
segment_sales = col.groupby(['Segment','Year'])['Sales'].sum().reset_index()
sns.barplot(x = 'Year', y = 'Sales', hue = 'Segment', data = segment_sales, palette = 'flare')
plt.title('Sales Trends for different Segments')
plt.show()
```

```python
# Product-wise sales analysis

col = ['Product Name', 'Sales']
col = store_df[col]

product_sales = col.groupby('Product Name')['Sales'].sum().reset_index()
product_sales = product_sales.sort_values(by='Sales', ascending = False)

barplot = sns.barplot(x = 'Sales', y = 'Product Name', data = product_sales.head(10),  palette = 'magma')
plt.title('Top 10 Products By Sales')

for index, value in enumerate(product_sales['Sales'].head(10)):
  barplot.text(value, index, f'{value:,.2f}', ha = 'right', va = 'center', color='white')
```
