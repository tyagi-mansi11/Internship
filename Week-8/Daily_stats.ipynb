```python
# Loading dataset
first_df = pd.read_csv('/content/daily_stats_data.csv')
df = first_df.copy()
```

# **Data Visualization**

---



```python
# Bar plot

fig = px.bar(df, x='activity_date', y='daily_average_heartrate', title='Relationship between daily activity and heart rate',
labels={'activity_date': 'Activity Date',
        'daily_average_heartrate': 'Daily Average Heartrate'},
        color_discrete_sequence=['yellowgreen'])
fig.show()
```

```python
# Bar plot

fig = px.bar(df, x='activity_date', y='total_steps', title='Relationship between daily activity and total steps taken',
labels={'activity_date': 'Activity Date',
        'total_steps': 'Total Steps'},
        color_discrete_sequence=['blueviolet'])
fig.show()
```

```python
# Scatter Plot

fig1 = px.scatter(df, x='total_steps', y='total_distance', title='Relationship between movement and measurement',
                  labels={'Total_steps', 'Total_distance'},
                  color_discrete_sequence=['slateblue'])
fig1.show()
```

```python
# Pair Plot
fig = px.scatter_matrix(df[['total_steps', 'total_distance','fairly_active_minutes']],
                        title='Daily Fitness Snapshot',
                        labels={'total_steps':'Total Steps', 'total_distance':'Total Distance', 'fairly_active_minutes':'Fairly active mins'},
                        color_discrete_sequence=['sandybrown'])
fig.show()
```

```python
# Stacked bar chat

fig = px.bar(df, x = 'activity_date', y=['fairly_active_minutes', 'very_active_minutes', 'lightly_active_minutes'],
             title='Comparing low, moderate, and high-intensity activity')
fig.show()
```

```python
 # Correlation Matrix

cols = ['total_steps', 'total_distance', 'tracker_distance', 'logged_activities_distance', 'sedentary_active_distance']
first_df = df[cols]
matrix = first_df.corr()
fig = px.imshow(matrix, labels=dict(color="Correlation"), x = matrix.index, y = matrix.columns, color_continuous_scale = 'matter')
fig.update_layout(title_text='Analysis of Step Count and Distance Correlation via a Correlation Matrix')
fig.show()
```
