```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Loading dataset
first_df = pd.read_csv('/content/minute_stats_data.csv')
df = first_df.copy()
```

# **Data Visualization**

---



```python
# Bar plot

fig = px.bar(df, x = 'activity_day', y = 'steps', title='Relationship between activity per minute and steps',
labels={'activity_day': 'Activity minute',
        'steps': 'Total steps'},
        color_discrete_sequence=['darkblue'])
fig.show()
```

```python
# Scatter Plot

fig1 = px.scatter(df, x='mets', y='calories', title='Relationship between mets and calories',
                  labels={'Total_steps', 'Calories'},
                  color_discrete_sequence=['darkblue'])
fig1.show()
```

```python
# Pair Plot
fig = px.scatter_matrix(df[['intensity', 'sleep','calories']],
                        title='Minute wise Fitness Snapshot',
                        labels={'sleep':'Sleep', 'intensity':'Intensity', 'calories':'Calories'},
                        color_discrete_sequence=['maroon'])
fig.show()
```
